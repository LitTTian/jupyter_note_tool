### tikz_to_image
- 参数:
  - `tikz_code`: tikz代码
  - `border`: 图片边界的宽度
  - `filename`: 保存的文件名
  - `output_dir`: 保存的目录
  - `dpi`: 图片分辨率
  - `toSave`: 是否保存图片
  - `toLog`: 是否打印`latex`日志
- 比如使用`tikz_to_image`函数绘制一个二维dp的过程图解
